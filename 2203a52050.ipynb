```python
data=[1,2,3,4,5]
print("List:",data)
print(type(data))
```

```python
array1= np.array(data)
array2=np.array([6,7,8,9,10])
print("\nArray 1:",array1)
print("Array 2:",array2)
print(type(array1))
```

```python
sin=np.sin(array1)
log=np.log(array1)
log2=np.log2(array1)
exp=np.exp(array1)
print("Sin Result:",sin)
print("Log Result:",log)
print("Log2 Result:",log2)
print("Exponential Result:",exp)
```
